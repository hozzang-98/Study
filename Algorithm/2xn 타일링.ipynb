```python
n = 4 

def solution(n):

    cnt = [0 for i in range(n)] #          - Memoization

    cnt[0] = 1 #  1 :    1
    cnt[1] = 2 #  2 : 0     + 1      1   
               #  3 :  1      + 2      ...
    
    #  
    # n  : n-2        + n-1       

    for i in range(2,n):
    
        cnt[i] = (cnt[i-2] + cnt[i-1]) % 1000000007
        
    return cnt[n-1]

solution(n)
```
