```python
book_time = [["15:00", "17:00"], ["16:40", "18:20"], ["14:20", "15:20"], ["14:10", "19:20"], ["18:20", "21:20"]]

def solution(book_time):

    from heapq import heappop, heappush

    #  ,  
    book_time = [[int(i[0].split(':')[0])*60 + int(i[0].split(':')[1]), int(i[1].split(':')[0])*60 + int(i[1].split(':')[1])] for i in book_time]
    book_time.sort()

    answer = 0

    #   
    heap = []

    #  10   
    for start_time, end_time in book_time:
        
        if not heap: #   Push

            heappush(heap, end_time+10)
            answer += 1
            continue

        if heap[0] <= start_time: #     heap        

            heappop(heap)
            # heappush(heap, end_time+10)
        else: #       

            answer += 1
            # heappush(heap, end_time+10)

        heappush(heap, end_time+10)

    return answer
        
solution(book_time)
```
