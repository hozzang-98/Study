```python
str1 = "FRANCE"
str2 = "FRENCH"

def solution(str1, str2):
    
    import re
    #      
    str1_list = [str1[i:i+2].lower() for i in range(len(str1)-1) if not re.findall('[^a-zA-Z]+', str1[i:i+2])]
    str2_list = [str2[i:i+2].lower() for i in range(len(str2)-1) if not re.findall('[^a-zA-Z]+', str2[i:i+2])]

    #   
    intersection = set(str1_list)&set(str2_list)
    union = set(str1_list)|set(str2_list)

    #        Jacard  
    # 
    jacard_intersection = sum([min(str1_list.count(i), str2_list.count(i)) for i in intersection]) 
    # 
    jacard_union = sum([max(str1_list.count(i), str2_list.count(i)) for i in union])

    if jacard_union == 0: 
        return 65536
    else: 
        return int(jacard_intersection/jacard_union * 65536)

solution(str1, str2)
```
