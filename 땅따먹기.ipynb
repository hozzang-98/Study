```python
land = [[1,2,3,5],[5,6,7,8],[4,3,2,1]]

def solution(land):

    row_num = len(land)
    col_num = len(land[0])

    for i in range(1,row_num):

        for j in range(col_num):

            land[i][j] = land[i][j] + max(land[i-1][:j] + land[i-1][j+1:])


    answer = max(land[-1])

    return answer

solution(land)
```
