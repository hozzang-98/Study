```python
n, m = map(int, input().split())   #    (n)   (m)
lan_list = []   #      
for _ in range(n):
    lan_list.append(int(input()))   #       .

start = 1
end = max(lan_list)

while start<=end:

    mid = (start+end) // 2

    cnt = 0

    for lan in lan_list: 
        
        cnt += lan//mid

    if cnt < m:

        end = mid - 1

    else:

        start = mid + 1
```
