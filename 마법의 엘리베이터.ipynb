```python
storey = 555

# +- 10 0,1,2,3..     
#     storey  0  
# 1.   6 ~ 9   10      .
# 2.   0 ~ 4   0      .
# 3.   5          .
# 3-1.    5 ~ 9    10      .
# 3-2.    0 ~ 4    0      .

def solution(storey):

    magic_stone = 0

    while storey:

        mod = storey % 10 

        if mod >= 6 or (mod == 5 and storey//10%10 >= 5):  # 1, 3-1

            count = 10 - mod 
            
            magic_stone += count
            storey += count

        else: # 2, 3-2

            count = mod

            magic_stone += count
            storey -= count

        storey //= 10

    return magic_stone

solution(storey)
```
