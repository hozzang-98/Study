```python
n = 5

arr1 = [9, 20, 28, 18, 11]

arr2 = [30, 1, 21, 17, 28]

def solution(n, arr1, arr2):
    
    answer = []

    for a, b in zip(arr1, arr2):

        binary_or = bin(a|b) #   OR

        oz = binary_or.replace('0b','').rjust(n,'0')
        ans = oz.replace('1','#').replace('0',' ')

        answer.append(ans)
    
    return answer
  
solution(n, arr1, arr2)
```
