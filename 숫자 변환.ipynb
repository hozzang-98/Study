```python
x = 10
y = 40
n = 5

def solution(x, y, n):
    
    cnt = 0
    
    now_level = set()
    now_level.add(x)
    
    while now_level:
        
        if y in now_level: #      y  answer
            
            return cnt
        
        else:
            
            next_level = set() #     ,     Set  
            
            for node in now_level:
                
                if node + n <= y: #  1
                    
                    next_level.add(node + n)
                    
                if node * 2 <= y: #  2
                    
                    next_level.add(node * 2)
                    
                if node * 3 <= y: #  3
                    
                    next_level.add(node * 3)
                
            now_level = next_level 
            cnt += 1 #  Level     1 
    
    return -1

solution(x, y, n)
```
