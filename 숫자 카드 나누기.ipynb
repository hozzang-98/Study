```python
arrayA = [14, 35, 119]
arrayB = [18, 30, 102]

def solution(arrayA, arrayB):

    import math

    gcdA, gcdB = 0, 0

    for a, b in zip(arrayA, arrayB):

        gcdA = math.gcd(gcdA, a) #  1-1:         
        gcdB = math.gcd(gcdB, b) #  2-2:         


    for a, b in zip(arrayA, arrayB):

        if gcdA != 0 and b % gcdA == 0: #  1-2:        1-2 

            gcdA = 0

        if gcdB != 0 and a % gcdB == 0: #  2-2:        2-2 
            
            gcdB = 0

    
    answer = max(gcdA, gcdB)

    return answer

solution(arrayA, arrayB)
```
