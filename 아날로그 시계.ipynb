```python
def solution(h1, m1, s1, h2, m2, s2):

    answer = 0

    startTime = h1 * 60 * 60 + m1 * 60 + s1

    endTime = h2 * 60 * 60 + m2 * 60 + s2

    if startTime == 0 or startTime == 12 * 3600:

        answer += 1

    while startTime < endTime:


        h_current_angle = startTime * 1 / 120 % 360
        m_current_angle = startTime * 0.1 % 360
        s_current_angle = startTime * 6 % 360

        if (startTime + 1) * 1/120 % 360 == 0:
            
            h_next_angle = 360
            
        else:
            
            h_next_angle = (startTime + 1) * 1/120 % 360
            
        if (startTime + 1) * 0.1 % 360 == 0:
            
            m_next_angle = 360
            
        else:
            
            m_next_angle = (startTime + 1) * 0.1 % 360
            
        if (startTime + 1) * 6 % 360 == 0:
            
            s_next_angle = 360
            
        else:
            
            s_next_angle = (startTime + 1) * 6 % 360
        

        if s_current_angle < h_current_angle & s_next_angle >= h_next_angle:

            answer += 1

        if s_current_angle < m_current_angle & s_next_angle >= m_next_angle:

            answer += 1

        if s_next_angle == h_next_angle & s_next_angle == m_next_angle:

            answer -= 1

        startTime += 1


    return answer
```
