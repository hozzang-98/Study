```python
n = 4 

def solution(n):

    cnt = [0 for i in range(n)] # 이미 계산된 함수 값을 저장해두기 위해 해당 리스트 생성 - Memoization

    cnt[0] = 1 # 가로가 1인 경우: 세로 배치하는 방법 1개
    cnt[1] = 2 # 가로가 2인 경우: 0인 경우 가로로 추가하는 방법 + 1인 경우에 사용한 세로 배치하는 방법 1개에다가 세로로 추가하는 방법
               # 가로가 3인 경우: 가로가 1인 경우에 가로로 추가 배치하는 방법 + 2인 경우에 세로로 추가 배치하는 방법 ...
    
    # 동적 계획법
    # n의 경우의 수: n-2의 모든 경우에 가로로 타일을 배치하는 경우의 수 + n-1이었던 모든 경우에 세로로 타일을 배치하는 경우의 수

    for i in range(2,n):
    
        cnt[i] = (cnt[i-2] + cnt[i-1]) % 1000000007
        
    return cnt[n-1]

solution(n)
```
