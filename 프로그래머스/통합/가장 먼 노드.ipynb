n개의 노드가 있는 그래프가 있습니다. 

각 노드는 1부터 n까지 번호가 적혀있습니다. 

1번 노드에서 가장 멀리 떨어진 노드의 갯수를 구하려고 합니다. 

가장 멀리 떨어진 노드란 최단경로로 이동했을 때 간선의 개수가 가장 많은 노드들을 의미합니다.

노드의 개수 n, 간선에 대한 정보가 담긴 2차원 배열 vertex가 매개변수로 주어질 때, 1번 노드로부터 가장 멀리 떨어진 노드가 몇 개인지를 return 하도록 solution 함수를 작성해주세요.

```python
n = 6
vertex = [[3, 6], [4, 3], [3, 2], [1, 3], [1, 2], [2, 4], [5, 2]]

def solution(n, vertex):

    from collections import deque

    route = [-1 for _ in range(n)] # Default -1, 간 곳은 노드1과의 거리를 설정할 리스트
    route[0] = 0 # 노드1 자신과의 거리는 0 으로 설정
    graph = [[] for i in range(n)]  # 노드 별 자신과 인접한 노드 정보 리스트 

    dq = deque([1]) # 출발 노드 설정된 Queue

    for s, e in vertex: # 양방향 그래프 설정

        graph[s-1].append(e)
        graph[e-1].append(s)

    while dq:

        node = dq.popleft()

        for n in graph[node-1]:

            if route[n-1] == -1:

                dq.append(n)
                route[n-1] = route[node-1] + 1 # 거리 1씩 추가

    max_val = max(route)
    answer = route.count(max_val)

    return answer

solution(n, vertex)
```
