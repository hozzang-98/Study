우리는 5g 기지국을 최소로 설치하면서 모든 아파트에 전파를 전달하려고 합니다. 

위의 예시에선 최소 3개의 아파트 옥상에 기지국을 설치해야 모든 아파트에 전파를 전달할 수 있습니다.


아파트의 개수 N, 현재 기지국이 설치된 아파트의 번호가 담긴 1차원 배열 stations, 전파의 도달 거리 W가 매개변수로 주어질 때, 

모든 아파트에 전파를 전달하기 위해 증설해야 할 기지국 개수의 최솟값을 리턴하는 solution 함수를 완성해주세요

```python
n = 11
stations = [4,11]
w = 1

def solution(n, stations, w):

    import math

    range = w * 2 + 1
    answer = 0
    start = 1

    for station in stations: 
        
        end = station - w

        if end < 2: # 맨 처음 아파트인 경우

            start = station + w + 1
        else: 
            apart_num = end - start
            answer += math.ceil(apart_num / range)
            start = station + w + 1

    # 맨 뒤 기치국 설치 아파트보다 뒤에 있는 경우 처리
    if station + w  < n:

        start = station + w
        apart_num = n -start

        answer += math.ceil(apart_num / range)

    return answer

solution(n,stations,w)
```
