```python
land = [[1,2,3,5],[5,6,7,8],[4,3,2,1]]

def solution(land):

    row_num = len(land)
    col_num = len(land[0])

    for i in range(1,row_num): # 첫 줄을 제외하고 한 줄씩 아래로

        for j in range(col_num):
            
            # 같은 열로는 내려갈 수 없으니 윗 줄의 :j, j+1: 열 중 최대 값을 더함 
            land[i][j] = land[i][j] + max(land[i-1][:j] + land[i-1][j+1:])


    answer = max(land[-1])

    return answer

solution(land)
```
