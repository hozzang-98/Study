```python
n, m = map(int, input().split())   # 가지고 있는 랜선의 개수(n)와 필요한 랜선의 개수(m)
lan_list = []   # 가지고 있는 랜선의 길이를 저장할 리스트
for _ in range(n):
    lan_list.append(int(input()))   # 가지고 있는 랜선의 길이를 입력받아 리스트에 추가합니다.

start = 1
end = max(lan_list)

while start<=end:

    mid = (start+end) // 2

    cnt = 0

    for lan in lan_list: 
        
        cnt += lan//mid

    if cnt < m: # 개수가 부족하면 길이 줄이기

        end = mid - 1

    else: # 개수가 많으면 길이 늘리기

        start = mid + 1
```
