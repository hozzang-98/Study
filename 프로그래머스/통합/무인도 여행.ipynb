```python
maps = ["X591X","X1X5X","X231X","1XXX1"]
    
def solution(maps):

    answer = []

    from collections import deque

    row, col = len(maps), len(maps[0])

    move = [(0,1), (1,0), (-1,0), (0,-1)]

    visited = [[False]*col for _ in range(row)]

    for i in range(row):

        for j in range(col):

            if maps[i][j] == "X" or visited[i][j] == True: # X거나 이미 간 곳인 경우 Continue
                
                continue

            q = deque()
            q.append((i,j)) # BFS 시작 지점
            visited[i][j] = True

            part_sum = 0

            while q:

                cx, cy = q.popleft()
                part_sum += int(maps[cx][cy])

                for dx, dy in move:
                    
                    # 오른쪽 이동 가능 여부 / 아래쪽 이동 가능 여부 / 바다 여부 / 기존 방문 여부
                    if (0 <= cx + dx < row) and (0 <= cy + dy < col) and (maps[cx+dx][cy+dy] != "X") and (visited[cx+dx][cy+dy] == False):

                        q.append((cx+dx, cy+dy))
                        visited[cx+dx][cy+dy] = True
            
            if part_sum != 0:

                answer.append(part_sum)


    if len(answer) != 0:

        answer.sort() # 오름차순

    else: # 아무 곳도 갈 수 없는 경우

        answer = [-1]

    return answer

solution(maps)
```
