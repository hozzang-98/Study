```python
N = 5
road = [[1,2,1],[2,3,3],[5,2,2],[1,4,2],[5,3,1],[5,4,2]]
k = 3

def solution(N, road, k):

    from collections import deque

    routes = [{} for _ in range(N+1)] # 노드마다 경로 딕셔너리 생성 
    distance = [k] * (N+1)
    distance[1] = 0 # 3, 0, 3, 3, 3, 3 
    available = {1}

    for sp, ep, d in road:

        if ep not in routes[sp] or routes[sp][ep] > d:

            routes[sp][ep] = d
            routes[ep][sp] = d

    dq = deque([1])

    while dq:

        start_point = dq.popleft()

        for end_point, dist in routes[start_point].items():
            
            if distance[end_point] >= distance[start_point] + dist: # 거리 k 안에 갈 수 있는 경우
        
                distance[end_point] = distance[start_point] + dist # 소모한 거리 갱신
            
                dq.append(end_point)
                available.add(end_point)
 
    return len(available)

solution(N, road, k)
```
