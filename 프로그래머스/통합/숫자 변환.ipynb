```python
x = 10
y = 40
n = 5

def solution(x, y, n):
    
    cnt = 0
    
    now_level = set()
    now_level.add(x)
    
    while now_level:
        
        if y in now_level: # 해당 연산을 완료한 결과 중 y가 있다면 answer
            
            return cnt
        
        else:
            
            next_level = set() # 연산들을 완료한 결과 값이 저장되며, 중복 값을 제거하기 위해 Set 자료형 사용
            
            for node in now_level:
                
                if node + n <= y: # 연산 1
                    
                    next_level.add(node + n)
                    
                if node * 2 <= y: # 연산 2
                    
                    next_level.add(node * 2)
                    
                if node * 3 <= y: # 연산 3
                    
                    next_level.add(node * 3)
                
            now_level = next_level 
            cnt += 1 # 하나의 Level 지날 시 연산 횟수 1 증가
    
    return -1

solution(x, y, n)
```
