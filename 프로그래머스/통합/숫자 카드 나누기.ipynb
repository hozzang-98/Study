```python
arrayA = [14, 35, 119]
arrayB = [18, 30, 102]

def solution(arrayA, arrayB):

    import math

    gcdA, gcdB = 0, 0

    for a, b in zip(arrayA, arrayB):

        gcdA = math.gcd(gcdA, a) # 조건 1-1: 철수 카드를 모두 나눌 수 있는 가장 큰 수
        gcdB = math.gcd(gcdB, b) # 조건 2-2: 영희 카드를 모두 나눌 수 있는 가장 큰 수


    for a, b in zip(arrayA, arrayB):

        if gcdA != 0 and b % gcdA == 0: # 조건 1-2: 영희 카드의 최대공약수를 나눌 수 있다면 조건 1-2 불만족

            gcdA = 0

        if gcdB != 0 and a % gcdB == 0: # 조건 2-2: 철수 카드의 최대공약수를 나눌 수 있다면 조건 2-2 불만족
            
            gcdB = 0

    
    answer = max(gcdA, gcdB)

    return answer

solution(arrayA, arrayB)
```
