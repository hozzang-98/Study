N개의 스티커가 원형으로 연결되어 있습니다. 다음 그림은 N = 8인 경우의 예시입니다.

원형으로 연결된 스티커에서 몇 장의 스티커를 뜯어내어 뜯어낸 스티커에 적힌 숫자의 합이 최대가 되도록 하고 싶습니다. 

단 스티커 한 장을 뜯어내면 양쪽으로 인접해있는 스티커는 찢어져서 사용할 수 없게 됩니다.

```python
stickers = [1, 3, 2, 5, 4]

def solution(stickers):
    # return 36
    length = len(stickers)

    # 스티커 개수가 하나면 해당 값 반환
    if length == 1: return stickers[0]
    
    # 스티커 개수가 2개면 둘 중 하나 큰거
    elif length == 2:

        return max(stickers)

    # 첫 장 포함하는 경우와 안 포함하는 경우에 대한 Dynamic Programming

    # 첫 장 포함
    pointer = [0] * length

    pointer[0] = stickers[0] # 첫장O

    pointer[1] = pointer[0] # 첫 장을 고르면 두 번째 장도 고를 수 없으므로 첫 장 포함 시 두 번째 값은 첫 장과 동일

    for i in range(2, length-1): # 첫 장 고르면 마지막 스티커는 못 뽑으니 Length - 1

        # print(pointer[i-1], pointer[i-2],stickers[i])
        pointer[i] = max(pointer[i-1], pointer[i-2] + stickers[i])  # (안 고르고 전 값 그대로 냅두기) vs (전전값 + 이번 스티커 고르기)

    max_1 = max(pointer)

    # 첫 장 미포함
    pointer = [0] * length

    pointer[1] = stickers[1] # 첫 번째 값은 0, 두 번째 값은 두 번째 장 고른거

    for i in range(2, length):

        # print(pointer[i-1], pointer[i-2],stickers[i])
        pointer[i] = max(pointer[i-1], pointer[i-2] + stickers[i])  # (안 고르고 전 값 그대로 냅두기) vs (전전값 + 이번 스티커 고르기)

    max_2 = max(pointer)

    return max(max_1, max_2)

solution(stickers)
```
