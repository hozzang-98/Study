```python
weights = [100,180,360,100,270]

def solution(weights):
    
    from collections import Counter

    answer = 0
    dic = Counter(weights)

    # 1:1 거리로 앉으면 균형인 경우 (서로 무게가 같은 경우) = v명일 때 v명 중 2명을 고르는 경우의 수
    for k, v in dic.items():

        if v >= 2:

            answer +=  v * (v-1) / 2 


    # 2:3, 2:4, 3:4
    for k, v in dic.items():

        if k * 2 / 3 in dic.keys():

            answer += v * dic[k * 2 / 3]

        if k * 2 / 4 in dic.keys():

            answer += v * dic[k * 2 / 4]

        if k * 3 / 4 in dic.keys():

            answer += v * dic[k * 3 / 4]

    return answer

solution(weights)
```
