n명의 사람이 일렬로 줄을 서고 있습니다. 

n명의 사람들에게는 각각 1번부터 n번까지 번호가 매겨져 있습니다. 

n명이 사람을 줄을 서는 방법은 여러가지 방법이 있습니다. 

예를 들어서 3명의 사람이 있다면 다음과 같이 6개의 방법이 있습니다.

1. [1, 2, 3]
2. [1, 3, 2]
3. [2, 1, 3]
4. [2, 3, 1]
5. [3, 1, 2]
6. [3, 2, 1]

사람의 수 n과, 자연수 k가 주어질 때, 사람을 나열 하는 방법을 사전 순으로 나열 했을 때, k번째 방법을 return하는 solution 함수를 완성해주세요.

```python
n = 3
k =  5

def solution(n, k):

    import math

    answer = []
    n_list = [i for i in range(1, n+1)]

    k -= 1 # 인덱스로 활용하기 위해 -1

    while n > 0:
    
        n -= 1 # 다음 자리수 생각

        fact = math.factorial(n) # 총 N개의 자리수 중에서. 첫번째 자리는 숫자 하나당 (N-1)!개 씩 존재, 다음 자리는 (N-2)!개 
        idx = k // fact # 앞자리가 정해진 경우 fact개만큼 존재하는데 그만큼 건너뛰면 됨
        answer.append(n_list.pop(idx)) # 건너뛰고 다음 숫자가 올 차례
      
        k %= fact # 나머지에 대해 동일한 과정 시행

    return answer

solution(n, k)
```
