```python
order = [4, 3, 1, 2, 5]

def solution(order):
    
    from collections import deque

    sub_con = deque()
    answer = 0
    idx = 1

    while idx <= len(order): 

        sub_con.append(idx) # 보조컨테이너에 idx번 상자부터 차례로 추가

        while sub_con and sub_con[-1] == order[answer]: # 보조컨테이너에 마지막에 있는 상자가 빼야되는 순서라면 택배를 정리 / 아니라면 (idx+1)번 택배로

            answer += 1
            sub_con.pop() # 택배 정리

        idx += 1 # 1, 2, 3, 4, 5

    return answer

solution(order)
```
