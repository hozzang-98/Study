```python
order = [4, 3, 1, 2, 5]

def solution(order):
    
    from collections import deque

    sub_con = deque()
    answer = 0
    idx = 1

    while idx <= len(order): 

        sub_con.append(idx) #  idx   

        while sub_con and sub_con[-1] == order[answer]: #         /  (idx+1) 

            answer += 1
            sub_con.pop() #  

        idx += 1 # 1, 2, 3, 4, 5

    return answer

solution(order)
```
