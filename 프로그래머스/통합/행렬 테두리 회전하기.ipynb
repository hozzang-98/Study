```python
rows = 6
columns = 6
queries = [[2,2,5,4],[3,3,6,6],[5,1,6,3]]

def solution(rows,columns,queries):

    # matrix
    matrix = [[columns*j+(i+1)for i in range(columns)] for j in range(rows)]

    answer = []

    for row1, col1, row2, col2 in queries:

        tmp = matrix[row1-1][col1-1]
        current_min = tmp

        # print("왼쪽 하->상") / 
        for k in range(row1-1, row2-1): #열 고정, 행 이동

            start = matrix[k+1][col1-1]
            matrix[k][col1-1] = start

            current_min = min(start, current_min)

        # print("하단 우->좌")
        for k in range(col1-1, col2-1): #행 고정, 열 이동

            start = matrix[row2-1][k+1]
            matrix[row2-1][k] = start

            current_min = min(start, current_min)

        # print("오른쪽 상->하")
        for k in range(row2-1, row1-1, -1): #열 고정, 행 이동

            start = matrix[k-1][col2-1]
            matrix[k][col2-1] = start

            current_min = min(start, current_min)

        # print("상단 좌->우")
        for k in range(col2-1, col1-1, -1): #행 고정, 열 이동

            start = matrix[row1-1][k-1]
            matrix[row1-1][k] = start

            current_min = min(start, current_min)

        # 첫 이동 값
        matrix[row1-1][col1] = tmp
        answer.append(current_min)
        
    return answer
    
solution(rows, columns, queries)
```
